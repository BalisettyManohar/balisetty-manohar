```python
x_features=list(df.columns)
x_features.remove('status')
encoded_df=pd.get_dummies(df[x_features],drop_first=True)
print(list(encoded_df.columns))
```

```python
x=encoded_df
y=df['status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='gini',max_depth=3)
clf.fit(x_train,y_train)
```

```python
pred_y=clf.predict(x_test)
```

```python
from sklearn import metrics
print("Cofusion Matrix is\n",metrics.confusion_matrix(pred_y,y_test))

print("Accuracy is",metrics.accuracy_score(y_test,pred_y))
print("AUS Score is",metrics.roc_auc_score(y_test,pred_y))
```

```python
from sklearn.tree import export_graphviz
import pydotplus as pdot
from IPython.display import Image
export_graphviz(clf,out_file="tree.odt",feature_names=x_train.columns,filled=True)
graph=pdot.graph_from_dot_file("tree.odt")
graph.write_png("tree.png")
Image(filename="tree.png")
```
